#### Creating Series

```python
# Creating Pandas Series

data = [10, 20, 30, 40]

series = pd.Series(data)

print(series)
```

```python
# Custom Index

series = pd.Series(data, index=['a', 'b', 'c', 'd'])

print(series)
```

#### Creating DataFrames

```python
# Creating DataFrame from Dictionary

data = {
    "Name": ["Div", "Aman", "Rahul", "Priya"],
    "Age": [22, 23, 21, 24],
    "City": ["Delhi", "Mumbai", "Jaipur", "Pune"]
}

df = pd.DataFrame(data)

df
```

```python
# DataFrame from List

data = [
    [1, "Laptop", 50000],
    [2, "Phone", 30000],
    [3, "Tablet", 20000]
]

df = pd.DataFrame(data, columns=["ID", "Product", "Price"])

df
```
