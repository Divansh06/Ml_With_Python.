##### Normalization

```python
arr = np.array([10, 20, 30, 40, 50], dtype=float)
```

```python
# min-max normalization to get the data in between 0 and 1
normalized = (arr-np.min(arr))/(np.max(arr)-np.min(arr))
normalized
```

```python
# Standardization (mean=0, std=1)
standardized = (arr - np.mean(arr)) / np.std(arr)
standardized
```

##### Dot product and Matrix multiplication

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

np.dot(a, b)       # 32 → (1×4 + 2×5 + 3×6)
```

```python
# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(A @ B)  
```

##### Comparison and Boolean operations

```python
arr1 = np.array([1, 5, 3, 8, 2, 9])
arr1 > 4         # [F T F T F T] → boolean array
```

```python
print(arr1[arr1 > 4])      # [5 8 9] → only those  elements which satisfy the condition ( only true elements)
```

```python
# use in ML
predictions = np.array([0.2, 0.8, 0.6, 0.3])
labels = (predictions > 0.5).astype(int)
labels  # [0 1 1 0] → thresholding!
```
