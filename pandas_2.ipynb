##### 7. Filtering Data

```python
# Age greater than 22

df[df["Age"] > 22]
```

```python
# Multiple Conditions

df[(df["Age"] > 21) & (df["City"] == "Pune")]
```

```python
# Using isin()

df[df["City"].isin(["Delhi", "Pune"])]
```
