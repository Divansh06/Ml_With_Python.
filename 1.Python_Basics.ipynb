## Strings and Bool

```python
name = "Divanshu123"
print(name.isalnum())
name1 = "Divanshu_123"
print(name1.isalnum())
name2 = ""

# in name we got "True" becoz the name consist of alphabet or numeric only which is what alnum checks .
# in name1 we got "false" becoz the name consist of alphabet , special charachter and numeric, so special charachter breaks the rule .
```

```python
print(name.istitle())
# should have the an alphabet in title case.
```

```python
print(name.isalpha())
```

```python
print(name.isupper()) # all charachter are uppercase
print(name.islower()) # all are lowercase
print(name.startswith('D')) 
print(name.endswith('3'))
```
